```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import pandas as pd
import seaborn as sns
import scipy.stats
```

```python
polski2015 = pd.read_csv('polski2015.csv')
polski2023 = pd.read_csv('polski2023.csv')
```

```python
#dane 2015
sredni_wynik2015 = polski2015['średni wynik (%)']
uczniowie2015 = polski2015['* liczba zdających']
RSPO_2015 = polski2015['RSPO szkoły']
zdawalnosc_2015 = polski2015['* zdawalność (%).1']

#dane 2023
sredni_wynik2023 = polski2023['średni wynik (%)']
uczniowie2023 = polski2023['* liczba zdających']
RSPO_2023 = polski2023['RSPO szkoły']
zdawalnosc_2023 = polski2023['* zdawalność (%).1']
```

```python
colors2015 = ['red' if x <= 30 else 'orange' if x <= 50 else 'yellow' if x <= 70 else 'green' for x in sredni_wynik2015]
colors2023 = ['red' if x <= 30 else 'orange' if x <= 50 else 'yellow' if x <= 70 else 'green' for x in sredni_wynik2023]


legend_patches = [
    Patch(color='red', label='wynik <= 30%'),
    Patch(color='orange', label='30% < wynik <=50%'),
    Patch(color='yellow', label='50% < wynik <= 70%'),
    Patch(color='green', label='70% < wynik')
]


plt.figure(figsize=(16, 6))
plt.suptitle('Wyniki egzaminu maturalnego z j. polskiego dla formuł 2015 i 2023', fontsize=16)

# Dla danych z 2015 roku
plt.subplot(1,2,1)
plt.scatter(range(len(sredni_wynik2015)), sredni_wynik2015, color=colors2015, s=10)
plt.xlabel('Przypisany numer szkoły ')
plt.ylabel('Średni wynik (%)')
plt.title('Formuła 2015')
plt.legend(handles=legend_patches, loc='lower right') 

# Dla danych z 2023 roku
plt.subplot(1,2,2)
plt.scatter(range(len(sredni_wynik2023)), sredni_wynik2023, color=colors2023, s=10)
plt.xlabel('Przypisany numer szkoły')
plt.ylabel('Średni wynik (%)')
plt.title('Formuła 2023')
plt.legend(handles=legend_patches, loc='lower right')

plt.savefig('wynik_dla_szkol.png')
plt.show()
```

```python
plt.figure(figsize=(16, 6))
plt.suptitle('Wyniki egzaminu maturalnego z j. polskiego w stosunku do liczby zdających', fontsize=16)
# Dla danych z 2015 roku
plt.subplot(1,2,1)
plt.scatter(uczniowie2015, sredni_wynik2015, color=colors2015, s=10)
plt.xlabel('Liczba zdających')
plt.ylabel('Średni wynik (%)')
plt.title('Formuła 2015')
plt.legend(handles=legend_patches, loc='lower right') 

# Dla danych z 2023 roku
plt.subplot(1,2,2)
plt.scatter(uczniowie2023, sredni_wynik2023, color=colors2023, s=10)
plt.xlabel('Liczba zdających')
plt.ylabel('Średni wynik (%)')
plt.title('Formuła 2015')
plt.legend(handles=legend_patches, loc='lower right') 

plt.savefig('wynik_od_zdajacych.png')
plt.show()
```
